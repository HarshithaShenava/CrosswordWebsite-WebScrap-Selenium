```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

```python
df=pd.DataFrame(columns=['Titles'])
```

```python
driver=webdriver.Chrome()
driver.get('https://www.crossword.in/collections/books')
time.sleep(10)
author_elems = driver.find_elements(By.CLASS_NAME, 'product-item-author')
title_elems = driver.find_elements(By.CLASS_NAME, 'product-item-title')
# Add each element's text to the DataFrame
#For titles
for i in title_elems:
    length = len(df)
    df.loc[length] = [i.text]
#For authors
for index, elem in enumerate(author_elems):
    if index < len(df):
        df.at[index, 'Authors'] = elem.text

# Close the browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Initialize lists to store titles and authors
titles = []
authors = []

# Initialize the web driver
driver = webdriver.Chrome()
driver.get('https://www.crossword.in/collections/books')

# Allow some time for the page to load
time.sleep(10)

# Find all div elements with the class 'result-product-item-info'
product_info_divs = driver.find_elements(By.CLASS_NAME, 'result-product-item-info')

# Iterate through each product info div and extract title and author
for product_info_div in product_info_divs:
    title_elem = product_info_div.find_element(By.CLASS_NAME, 'product-item-title')
    titles.append(title_elem.text)
    
    try:
        author_elem = product_info_div.find_element(By.CLASS_NAME, 'product-item-author')
        authors.append(author_elem.text)
    except:
        authors.append('NA')

# Close the browser
driver.quit()

# Create a DataFrame from the lists
new_df = pd.DataFrame({'Title': titles, 'Author': authors})

```

```python
driver=webdriver.Chrome()
driver.get('https://www.crossword.in/collections/books')
price=driver.find_elements(By.CLASS_NAME,'wizzy-product-item-price')
for index, elem in enumerate(price):
    if index < len(new_df):
        new_df.at[index, 'Price'] = elem.text.split(' ')[0]
driver.close()
```
